```python
data = pd.read_excel('Данные для тестового задания.xlsx', sheet_name='Данные АБ тестов')
```

```python
from scipy import stats

for i in [1, 2, 3]:
    exp_data = data[data['experiment_num'] == i]
    group_a = exp_data[exp_data['experiment_group'] == 'control']['revenue']
    group_b = exp_data[exp_data['experiment_group'] == 'test']['revenue']
    
    
    arpu_a = group_a.mean()
    arpu_b = group_b.mean()
    lift = (arpu_b - arpu_a) / arpu_a
    
   
    t_stat, p_val = stats.ttest_ind(group_a, group_b)
    
    
    print(f"Эксперимент №{i}:")
    print(f"ARPU: Контроль = {arpu_a:.2f}, Тест = {arpu_b:.2f} (Прирост: {lift:.2%})")
    print(f"p-value = {p_val:.4f}")
    
```

```python
from scipy import stats

for i in [1, 2, 3]:
    exp_data = data[data['experiment_num'] == i]
    group_a = exp_data[exp_data['experiment_group'] == 'control']['revenue']
    group_b = exp_data[exp_data['experiment_group'] == 'test']['revenue']
    
    print(f"--- Эксперимент №{i} ---")
    
    for name, group in zip(['Контроль', 'Тест'], [group_a, group_b]):
        shapiro_stat, p_val = stats.shapiro(group)
        
        is_normal = "Нормальное" if p_val > 0.05 else "НЕ нормальное"
        
        print(f"{name}: p-value = {p_val:.10f} ({is_normal})")
    
    
```

```python
for i in [1, 2, 3]:
    exp_data = data[data['experiment_num'] == i]
    group_a = exp_data[exp_data['experiment_group'] == 'control']['revenue']
    group_b = exp_data[exp_data['experiment_group'] == 'test']['revenue']
    
    arpu_a = group_a.mean()
    arpu_b = group_b.mean()
    lift = (arpu_b - arpu_a) / arpu_a
    
    
    u_stat, p_val_mw = stats.mannwhitneyu(group_a, group_b, alternative='two-sided')
    
    
    print(f"Эксперимент №{i}:")
    print(f"U-stat: {u_stat:.2f}, p-value (MW): {p_val_mw:.4f}")
    print(f"Прирост ARPU: {lift:.2%}")
    
    
    if p_val_mw < 0.05:
        print("Результат: Различия в распределениях СТАТИСТИЧЕСКИ ЗНАЧИМЫ")
    else:
        print("Результат: Различия СЛУЧАЙНЫ (незначимы)")
    
```

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

visitors = np.array([100047501, 100001055]) 
payments = np.array([1003, 1099])         

cr_a = payments[0] / visitors[0]
cr_b = payments[1] / visitors[1]
lift = (cr_b - cr_a) / cr_a


z_stat, p_value = proportions_ztest(payments, visitors)

print(f"Конверсия A: {cr_a:.7%}")
print(f"Конверсия B: {cr_b:.7%}")
print(f"Относительный прирост (Lift): {lift:.2%}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Результат: СТАТИСТИЧЕСКИ ЗНАЧИМ")
else:
    print("Результат: НЕЗНАЧИМ")
```
